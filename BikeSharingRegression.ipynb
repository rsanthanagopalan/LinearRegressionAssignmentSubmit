## Splitting the dataset into train and test

## Rescaling features

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
num_vars = ['hum', 'temp', 'atemp', 'windspeed','cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

```python
df_train.describe()
```

## Building a linear model & we will start with a coarse grain approach using RFE

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, step=15)            
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
features = X_train.columns[rfe.support_]
features
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[features]
```

## Fine grain check on independent variables including VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
import statsmodels.api as sm

# Add a constant
X_train_lm1 = sm.add_constant(X_train_rfe)

lr1 = sm.OLS(y_train, X_train_lm1).fit()
```

```python
lr1.params
```

```python
lr1.summary()
```
